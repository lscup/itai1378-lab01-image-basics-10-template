# Image Basics 10

This notebook covers basic image manipulation tasks using popular libraries such as OpenCV and Matplotlib. You will learn how to perform image reading, conversion, and display using Python.

```python
# Task 1
# Import the required libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Print a message to confirm the imports worked
print("Libraries imported successfully!")
```

### Task 2

In this task, we need to load an image. We will simulate loading an image using a locally available image named `sample.jpg`, which is pre-seeded in the working directory.

```python
# Task 2
# Since internet access isn't allowed, we'll use the pre-seeded image.
file_name = 'sample.jpg'
image_bgr = cv2.imread(file_name)

# Check if the image is loaded correctly
if image_bgr is not None:
    print(f"Image '{file_name}' loaded successfully!")
else:
    print(f"Failed to load image '{file_name}'.")
```

### Task 3

Next, we read the image using OpenCV. It's crucial to check if the image was loaded correctly to avoid errors in later steps.

```python
# Task 3
# Check if the image is loaded correctly
type(image_bgr)
image_bgr.shape if image_bgr is not None else "No image loaded"
```

### Task 4

We will print the type and shape of the image to understand its dimensions and color channels.

```python
# Task 4
# Print the type of the image
print("Type:", type(image_bgr))

# Print the shape of the image
print("Shape:", image_bgr.shape)
```

### Task 5

OpenCV loads images in BGR format by default. We'll convert the image from BGR to RGB to display it correctly with Matplotlib, which expects RGB format.

```python
# Task 5
# Convert image_bgr to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

### Task 6

We'll display the image using Matplotlib. Remember to turn off the axes for a cleaner view and add a title.

```python
# Task 6
# Display the RGB image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off the axes
plt.title('RGB Image')
plt.show()
```

### Task 7

In this task, we will extract individual Red, Green, and Blue channels from the image and display each one separately using Matplotlib. This helps in analyzing the image data in each channel.

```python
# Task 7
# Extract Red, Green, and Blue channels from image_rgb
red_channel = image_rgb[:, :, 0]
green_channel = image_rgb[:, :, 1]
blue_channel = image_rgb[:, :, 2]

# Display each channel
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes[0].imshow(red_channel, cmap='Reds')
axes[0].set_title('Red Channel')
axes[0].axis('off')

axes[1].imshow(green_channel, cmap='Greens')
axes[1].set_title('Green Channel')
axes[1].axis('off')

axes[2].imshow(blue_channel, cmap='Blues')
axes[2].set_title('Blue Channel')
axes[2].axis('off')

plt.show()
```
